# **Validación de los modelos**

La construcción de un sistema de ML incluye la búsqueda de un algoritmo de aprendizaje óptimo y de los mejores hyperparámetros para el sistema, cuyo rendimiento se puede evaluar con un conjunto de validación como ya hemos visto.

Una pregunta clave es: ¿Si el modelo tiene un bajo rendimiento, qué se debería hacer? 

* ¿Usar un modelo más complejo/flexible?
* ¿Usar un modelo menos complejo/flexible?
* ¿Conseguir más datos de entrenamiento?

La respuesta a esta pregunta a menudo es contraintuitiva. En particular, a veces el uso de un modelo más complejo dará peores resultados, y agregar más muestras de entrenamiento puede que no mejore nuestros resultados

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```
def nl_data(N, seed=1):
  """genera datos con una relacion no lineal"""
  np.random.seed(seed)
  X = np.random.rand(N) ** 2
  y = 10 - 1/(X + 0.1) + 2*np.random.rand(N)

  return np.c_[X], y

X, y = nl_data(40)
```

```
def polyreg(degree=2):
  """retorna un modelo de regresion polinomial"""
  return make_pipeline(PolynomialFeatures(degree),
                       StandardScaler(),
                       LinearRegression())
```

```
def plot_polyreg(X, y, degree, seed=42):
  """grafica un modelo de regresion polinomial"""

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

  model = polyreg(degree).fit(X_train, y_train)

  x = np.c_[np.linspace(X.min(),X.max(),100)]
  y_pred = model.predict(x)

  print(f"train: {model.score(X_train, y_train)}")
  print(f"test: {model.score(X_test, y_test)}")
  
  fig = plt.figure(figsize=(10,7))
  plt.scatter(X_train, y_train, label="train data")
  plt.scatter(X_test, y_test, label="test data")
  plt.plot(x, y_pred, "r", label="model")
  plt.title(f"degree: {degree}")
  plt.legend();
```

## **Sesgo/varianza de los modelos**

Existen diferentes conceptos que se pueden utilizar para diagnosticar el comportamiento de un modelo de ML. Por ejemplo, los términos **sobreajuste** y **subajuste** se utilizan para diagnosticar el rendimiento de un modelo basado en los puntajes de entrenamiento y de prueba


**Subajuste**: Un modelo que sufre de *subajuste* tiene un bajo rendimiento sobre los datos de validación y de entrenamiento


```
plot_polyreg(X, y, 1)
```

En este caso se intenta ajustar una línea recta a los datos, pero debido a que estos son intrínsecamente más complicados que una línea recta, el modelo de línea recta nunca podrá describir bien este conjunto de datos. Se dice que tal modelo se "subajusta" los datos; es decir, no tiene suficiente flexibilidad para tener en cuenta adecuadamente todas las características de los datos

Posibles razones para el subajuste:

* El modelo es muy simple 
* Las características no son muy informativas
* Demasiada regularización 

Posibles soluciones al subajuste:

* Construir un modelo más complejo (añadir más parámetros)
* Transformar o aumentar las características para tener un poder predictivo mayor
* Añadir más datos de entrenamiento
* Reducir el ruido en los datos
* Reducir la regularización


**Sobreajuste**: Un modelo que sufre de *sobreajuste* tiene un buen rendimiento solo sobre los datos de entrenamiento 

```
plot_polyreg(X, y, 14)
```

Este modelo intenta ajustar un polinomio de orden superior a los datos. Aquí, el ajuste del modelo tiene suficiente flexibilidad para tener en cuenta casi perfectamente las características finas de los datos, pero aunque describe con mucha precisión los datos de entrenamiento, su forma precisa parece reflejar más las propiedades de ruido particulares de los datos que las propiedades intrínsecas de cualquiera sea el proceso con el que se generaron los datos. Se dice entonces que el modelo "sobreajusta" los datos.

Posibles razones para el sobreajuste:

* El modelo es muy complejo
* Demasiadas características y muy pocas instancias para el entrenamiento
* Poca regularización

Posibles soluciones al sobreajuste:

* Simplificar el modelo (reducir los parámetros)
* Simplificar los datos de entrenamiento (menos características)
* Añadir más datos de entrenamiento
* Construir un *ensemble*
* Regularizar el modelo


Otra forma de diagnosticar los modelos es utilizando los términos **sesgo** y **varianza**:

**Sesgo**: El sesgo en ML puede deberse a un error debido a suposiciones incorrectas en el algoritmo de aprendizaje. Un alto sesgo puede causar que un algoritmo no capture las relaciones relevantes entre las características y la(s) etiqueta(s) (subajuste). Una predicción sesgada puede ocurrir debido a datos "ruidosos", un conjunto de características incompleto o unas instancias de entrenamiento sesgadas.

El error debido al sesgo es la diferencia entre el valor esperado (media) de la predicción del modelo y el valor correcto que se quiere predecir:

$$\text{Bias} = E[\hat{y}] - y$$

El sesgo mide qué tan lejos están las predicciones de los valores correctos en general si reconstruimos el modelo varias veces en diferentes conjuntos de datos de entrenamiento. El sesgo es la medida del error sistemático que no se debe a la aleatoriedad.

**Varianza**: La varianza es el valor esperado de la desviación del cuadrado de la media


$$\text{Var} = E[\left(\hat{y} - E[\hat{y}]\right)^2]$$

Una alta varianza puede causar que un algoritmo modele el ruido aleatorio en el conjunto de entrenamiento en lugar de los etiquetas deseadas (sobreajuste)

El error debido a la varianza es la variabilidad de la predicción de un modelo para una instancia dada. La varianza mide la consistencia o variabilidad de la predicción del modelo para una instancia particular si volvemos a entrenar el modelo varias veces. Podemos decir que el modelo es sensible a la aleatoriedad de los datos de entrenamiento.



* Alto sesgo + baja varianza: subajuste
* Alta varianza + bajo sesgo: sobreajuste

## **Curvas de validación**

Al tratar de reducir la varianza, se incrementa el sesgo y viceversa. En otras palabras, reducir el sobreajuste conduce al subajuste, y viceversa. Esto se conoce como el *trade-off* del sesgo y la varianza: al tratar de construir un modelo que tenga un buen rendimiento sobre los datos de entrenamiento, terminamos con un modelo que tiene un mal rendimiento sobre los datos de validación. Básicamente, encontrar el "mejor modelo" es encontrar un punto óptimo de equilibrio entre el sesgo y la varianza. 

Muchos factores determinan el nivel de subajuste o sobreajuste de un modelo, pero el factor más importante es la complejidad del modelo. Si imaginamos que tenemos cierta capacidad para ajustar la complejidad del modelo, esperamos que los puntajes de entrenamiento y validación se comporten de la siguiente manera:

Esta visualización se conoce como la **curva de validación**. Para algún valor intermedio, la curva de validación tiene un máximo. Este nivel de complejidad indica un punto óptimo entre la varianza y el sesgo.

Los medios mediante los cuales podemos ajustar la complejidad del modelo varían de un modelo a otro; cuando analicemos modelos individuales en profundidad en secciones posteriores, veremos cómo cada modelo permite tal ajuste.

Podemos obtener estas curvas utilizando la clase `validation_curve` de Sklearn, la cual toma como argumentos principales el estimador a evaluar, los datos, el hiperparámetro que regula la complejidad (o en general algún parámetro para el cual queramos estudiar el comportamiento de los puntajes de entrenamiento y validación) y el rango en el que queremos explorar dicho hiperparámetro.

Recuerde que como tenemos nuestro estimador embebido en un pipeline, los nombres de los hiperparámetros se ven modificados. En general, estos tendrán la forma `estimador/transformador__hiperparámetro`. Para acceder directamente a esta información podemos utilizar el método `get_params`:

```
# hiperparametros del modelo
polyreg().get_params()
```

Construyamos entonces una curva de validación en función del grado de la regresión polinomial (`polynomialfeatures__degree`):

```
from sklearn.model_selection import validation_curve
 
def VC(X, y):
  degree_range = np.arange(1, 20)

  train_score, val_score = validation_curve(polyreg(), X, y, 
                                            param_name="polynomialfeatures__degree", 
                                            param_range=degree_range, cv=7)
  
  fig = plt.figure(figsize=(10,7))
  plt.plot(degree_range, np.median(train_score, axis=1), "b", label="train score")
  plt.plot(degree_range, np.median(val_score, axis=1), "r", label="validation score")
  plt.ylim(0, 1.1)
  plt.xlabel("degree")
  plt.ylabel("score")
  plt.legend();
```

```
VC(X, y)
```

Esto muestra precisamente el comportamiento cualitativo que esperamos: la puntuación de entrenamiento mejora monótonamente con una mayor complejidad del modelo; y la puntuación de validación alcanza un máximo antes de caer cuando el modelo se sobreajusta. De la curva de validación, podemos leer que el *trade-off* óptimo entre sesgo y varianza se encuentra para un polinomio de un grado alrededor de 4


```
plot_polyreg(X, y, 4)
```

Note que para encontrar este modelo óptimo no calculamos solamente el puntaje de entrenamiento sino que necesitamos examinar la relación entre los puntajes de entrenamiento y de validación. Esto puede brindarnos información útil sobre el rendimiento del modelo.

## **Curvas de aprendizaje**

Un aspecto importante de la complejidad del modelo es que el modelo óptimo generalmente dependerá del tamaño de los datos de entrenamiento.

Consideremos un conjunto de datos de 100 instancias y veamos cómo es su curva de validación:

```
X, y = nl_data(100)

VC(X, y)
```

De la curva de validación se desprende claramente que el conjunto de datos más grande puede admitir un modelo mucho más complejo: el pico aquí probablemente sea de alrededor de un grado 7-8, pero incluso un modelo de grado 15 no sobreajusta seriamente los datos; las puntuaciones de validación y entrenamiento permanecen muy cerca.

Vemos entonces que el comportamiento de la curva de validación no tiene una sino dos entradas importantes: la complejidad del modelo y el número de instancias de entrenamiento. 

Para explorar el comportamiento del modelo en función del número de instancias de entrenamiento podemos usar subconjuntos de datos cada vez más grandes para ajustar el modelo.  

Una gráfica de los puntajes de entrenamiento/validación con respecto al tamaño del conjunto de entrenamiento se conoce como *curva de aprendizaje*.

Esta curva nos pueda dar información respecto a si agregar más datos mejorará el desempeño del modelo. Podemos construir una curva de aprendizaje utilizando la función `learning_curve` de sklearn: 



```
from sklearn.model_selection import learning_curve

def LC(X, y, degree):
  N, train_score, val_score = learning_curve(polyreg(degree), X, y, cv=7, train_sizes=np.linspace(0.2, 1, 25))

  fig = plt.figure(figsize=(10,7))
  plt.plot(N, np.mean(train_score, axis=1), "b", label='training score')
  plt.plot(N, np.mean(val_score, axis=1), "r", label='validation score')
  plt.ylim(0, 1.01)
  plt.xlabel('training size')
  plt.ylabel('score')
  plt.title(f"degree: {degree}")
  plt.legend();
```

```
LC(X, y, 2)
```

Una vez que ambos puntajes convergen a un valor particular, agregar más datos de entrenamiento no será de ayuda. El puntaje de convergencia se podría mejorar si se entrena un modelo más complejo:

```
LC(X, y, 4)
```

Esta curva también nos puede informar si tenemos un problema de sobreajuste o subajuste. Dibujar una curva de aprendizaje para la elección particular de un modelo y un conjunto de datos puede ayudarnos a tomar decisiones sobre cómo avanzar en la mejora del análisis. 

Todo este análisis anterior nos da una idea del *trade-off* que existe entre el sesgo y la varianza, y su dependencia tanto con la complejidad del modelo como con el tamaño del conjunto de entrenamiento.

# **Ajuste de hiperparámetros**


Los hiperparámetros juegan un papel fundamental en el proceso de entrenamiento. Algunos hiperparámetros influencian la velocidad del entrenamiento, pero lo más importante es que los hiperparámetros controlan el *trade-off* de sesgo-varianza (y algunos otros *trade-offs* que aparecen en el ML)

Los hiperparámetros no se optimizan en el proceso de aprendizaje sino que son definidos previo a este proceso. Normalmente se experimenta con diferentes valores de los hiperparámetros. Sin embargo, cada modelo tiene un único conjunto de hiperparámetros. Más aún, cada etapa en la construcción del sistema de ML (preprocesado, entrenamiento y predicción) puede tener sus propios hiperparámetros.

Los resultados de la experimentación con diferentes valores de hiperparámetros podemos usarlos para mejorar el modelo final. El proceso es el siguiente:

* Entrenamos el modelo
* Evaluamos el modelo
* Ajustamos los hiperparámetros

y hacemos esto múltiples veces. ¿Qué pasará si hacemos esto muchas veces? Si repetimos este proceso muchas veces, el modelo empezará a mostrar muy buenos resultados en la evaluación y lentamente empezará a sobreajustar el conjunto de validación. En algún punto, obtendremos puntajes muy buenos que no representarán realmente el rendimiento del modelo.

La solución a esto es construir otro conjunto de validación: lo que se suele hacer es añadir el conjunto de validación al conjunto de prueba en cada iteración, y obtener un nuevo conjunto de validación. A este nuevo conjunto se le conoce como el "conjunto de prueba". Es decir, realizamos el proceso de entrenamiento, evaluación y ajuste de hiperparámetros con dos conjuntos de datos (entrenamiento + validación), y reservamos otro conjunto de datos para una evaluación más adecuada del modelo final.

La ventaja de esta estrategia es que el modelo nunca ha visto estos datos, ni directamente en el proceso de entrenamiento ni indirectamente en el proceso de validación. Esta es la mejor evaluación que podemos hacer para entender las capacidades reales del modelo.

## **GridSearch**

Esta es la técnica de ajuste de hiperparámetros más simple. Esta consiste en la discretización de cada uno de los hiperparámetros para posteriormente evaluar cada posible combinación de estos valores. Esta técnica se utiliza cuando el número de hiperparámetros y sus rangos no son muy grandes. 

Scikit-Learn proporciona herramientas automatizadas para hacer esto con el metaestimador `GridSearchCV`

Veamos un ejemplo del uso del *grid search* para encontrar el modelo polinomial óptimo. Exploraremos en concreto tres hiperparámetros del modelo: el grado del polinomio, el ajuste de la intersección (término de sesgo) y la normalización de las características. 


```
from sklearn.model_selection import GridSearchCV
```

Definimos los hiperparámetros y los rangos en que los queremos explorar: 

```
param_grid = {'polynomialfeatures__degree': np.arange(21),
                   'linearregression__fit_intercept': [True, False],
                   'linearregression__normalize': [True, False]}
```

Entrenamos el estimador

```
grid = GridSearchCV(polyreg(), param_grid, cv=5).fit(X, y)
```

El resultado del *grid search* se haya en el atributo `cv_results_`

```
grid.cv_results_.keys()
```

Visualicemos estos resultados en un dataframe:

```
results = pd.DataFrame({
    "degree":grid.cv_results_["param_polynomialfeatures__degree"],
    "normalize":grid.cv_results_["param_linearregression__normalize"],
    "intercept":grid.cv_results_["param_linearregression__fit_intercept"],
    "score":grid.cv_results_["mean_test_score"]})

results.sort_values(by=["score"], ascending=False).head()
```

Podemos accerder la información de los mejores hiperparámetros y su puntaje asociado mediante los atributos `best_params_` y `best_score_`, respectivamente.

```
# mejores parametros
grid.best_params_
```

```
# mejor puntaje
grid.best_score_
```

Adicionalmente, podemos seleccionar directamente un modelo que tenga estos hiperparámetros, el cual encontramos en el atributo `best_estimator_` 

```
# modelo con el mejor rendimiento
grid.best_estimator_
```

Si tenemos modelos muy complejos y un conjunto de datos de tamaño considerable, y realizamos el proceso de *grid search* sobre un rango amplio en los hiperparámetros, este proceso será muy costoso computacionalmente.

## **RandomSearch**

Esta técnica difiere del *grid search* en que no se provee una serie de valores discretos para explorar los hiperparámetros sino que se provee una distribución estadística (contínua o discreta de acuerdo a la naturaleza del hiperparámetro) para cada hiperparámetro a partir de la cual se muestrean los valores aleatoriamente.

Para la implementación podemos utilizar la clase `RandomizedSearchCV` de Sklearn

```
from sklearn.model_selection import RandomizedSearchCV

param_dist = {'polynomialfeatures__degree': np.arange(21),
                   'linearregression__fit_intercept': [True, False],
                   'linearregression__normalize': [True, False]}

grid = RandomizedSearchCV(polyreg(), param_dist, cv=5).fit(X, y)
```

```
grid.best_params_
```

```
grid.best_score_
```

```
grid.best_estimator_
```
