#  **Creando figuras y ejes**

Usaremos el submódulo `pyplot` de la librería para realizar nuestros gráficos:




```
import matplotlib.pyplot as plt
```

Existen dos paradigmas de visualización en matplotlib. La primera que vamos a ver se conoce como *interfaz tipo Matlab*, donde la figura y los ejes se crean automáticamente: 

Por ejemplo, cuando utilizamos la función `plot` directamente, que genera lo que conocemos como un gráfico de línea, se generan una figura y un conjunto de ejes de forma implícita

```
import numpy as np

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y);
```

La segunda forma utiliza un estilo de programación orientado a objetos: referenciamos la figura y los ejes mediante las variables `fig` y `ax` respectivamente, y serán métodos y atributos de estas objetos los que nos permitan personalizar la visualización. 

Una primera opción para crear la figura y los ejes es utilizar las funciones `plt.figure` y `plt.axes`, que nos devuelven una referencia de la figura y conjunto de ejes utilizados.

```
fig = plt.figure()
ax = plt.axes()
```

Una vez tenemos la referencia al conjunto de ejes `ax` podemos crear una gráfica ejecutando el método `plot` sobre `ax`:

```
# se crean la figura y los ejes
fig = plt.figure()
ax = plt.axes()

# se realiza la grafica mediante la referencia al eje
ax.plot(x, y);
```

Podemos utilizar alternativamente la función `subplots` para crear la figura y el conjunto de ejes:

```
# se crean la figura y los ejes
fig, ax = plt.subplots()

# se realiza la grafica mediante la referencia al eje
ax.plot(x, y);
```

Cuando utilizamos la interfaz tipo *matlab*, podemos obtener los objetos `Figure` y `Axes` (la figura y los ejes) con los métodos `gcf` y `gca` respectivamente

```
plt.plot(x, y)

fig = plt.gcf()
ax = plt.gca()
```

# **Personalización básica del gráfico**

* **Estilo MATLAB**

La función `plt.plot` realiza por defecto un gráfico de línea. Podemos personalizar el color, el estilo y el marcador de la línea mediante los argumentos por palabra clave `color`, `linestyle` y `marker`, respectivamente. E


```
plt.plot(x, y, color="black", linestyle="dashed", marker="o");
```

Acá podemos ver algunas opciones para los marcadores:
>   

    "o" ------ círculo
    "^" ------ triángulo
    "p" ------ pentágono
    "h" ------ hexágono
    "s" ------ cuadrado
    "|" ------ línea vertical
    "_" ------ línea horizontal
    "+" ------ más
    "x" ------ equis

Podemos pasar el estilo de la línea de forma completa o abreviada:

>
    COM        ABR      DESCRIPCIÓN
    solid ---- "-" ---- línea sólida
    dashed --- "--" --- línea discontinua
    dashdot -- "-." --- línea discontinua+punto
    dotted --- ":" ---- línea punteada  

De forma similar, en el caso del color podemos pasar este de forma completa o abreviada:

> 
    COM       ABR
    blue ---- b
    green --- g
    red ----- r
    cyan ---- c
    magenta - m
    yellow -- y  
    black --- k
    white --- w







La gráfica anterior la podríamos definir de forma alternativa como

```
plt.plot(x, y, color="k", linestyle="--", marker="o");
```

Podemos reducir estos tres argumentos a un solo `string` donde se indique (sin importar el orden) los tres caracteres correspondientes a los argumentos. Por ejemplo, con el string `"go:"`, se generará una línea de color verde, discontinua y con marcadores circulares:

```
plt.plot(x, y, "go:");
```

Podemos agregar al gráfico un título y etiquetas para los ejes $x$ y $y$, con los métodos `plt.title`, `plt.xlabel` y `plt.xylabel`, respectivamente:

```
plt.plot(x, y)

# etiquetas de los ejes
plt.xlabel("x")
plt.ylabel("sin(x)");

# titulo
plt.title("Funcion Seno");
```

Ahora, para definir los límites de los ejes tenemos dos opciones. La primera es utilizando los métodos `plt.xlim` y `plt.ylim`. Alternativamente, el método `plt.axis`, con el cual podemos especificar los límites de los ejes pasando una lista que contiene como elementos los límites de la siguiente manera:
>
    plt.axis([xmin, xmax, ymin, ymax])

```
plt.plot(x, y)

# etiquetas de los ejes
plt.xlabel("x")
plt.ylabel("sin(x)");

# titulo
plt.title("Funcion Seno")

# limites de los ejes
plt.axis([-2*np.pi, 2*np.pi, -1, 1]);
```

Con las palabras clave `tight` y `equal`, podemos definir los ejes de manera "ajustada" en el primer caso, y ejes iguales en el segundo caso (por defecto se utiliza `tight`)

```
theta = np.linspace(0, 2*np.pi)
X = np.cos(theta)
Y = np.sin(theta) 

plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y")

plt.axis("tight");
```

```
theta = np.linspace(0, 2*np.pi)
X = np.cos(theta)
Y = np.sin(theta) 

plt.plot(X, Y)
plt.xlabel("x")
plt.ylabel("y");

plt.axis("equal");
```

Para agregar una leyenda de la función, podemos pasar una lista con la(s) leyenda(s) a la función `plt.legend`

```
plt.plot(x, y)

plt.xlabel("x")
plt.ylabel("sin(x)");
plt.title("Funcion Seno")

# añadir leyenda
plt.legend(["sin(x)"]);
```

o utilizando el argumento por palabra clave ``label`` dentro de la función ``plt.plot()``

```
plt.plot(x, y, label="sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)");
plt.title("Funcion Seno")

# añadir leyenda
plt.legend();
```

Podemos además especificar, mediante el argumento por palabra clave `loc`, la posición de la leyenda dentro del gráfico. Por defecto, se utiliza la "mejor" posición (best). Las opciones disponibles son:
>
    POSICIÓN        DESCRIPCIÓN
    best ---------- mejor
    upper right --- sup der
    upper left ---- sup izq
    lower left ---- inf izq 
    lower right --- inf der
    right --------- der 
    center left --- centro izq
    center right -- cento der
    lower center -- inf central
    upper center -- sup central
    center -------- central

```
plt.plot(x, y, label="$sin(x)$")
plt.xlabel("x")
plt.ylabel("sin(x)");
plt.title("Funcion Seno")

# añadir leyenda
plt.legend(loc="lower left");
```

Nótese que debemos incluir `plt.legend()` para que se genere la leyenda. Además, podemos mostrar el título, los ejes y la leyenda, en un formato matemático encerrando la expresión entre los símbolos peso `$`

```
plt.plot(x, y, label="$sin(x)$")
plt.xlabel("$x$")
plt.ylabel("$sin(x)$");
plt.title("Funcion Seno")
plt.legend();
```

* **Estilo orientado a objetos**

Si bien la mayoría de las funciones en `plt` se traducen directamente a métodos `ax` (como `plt.plot()` → `ax.plot()`, `plt.legend()` → `ax.legend()`, etc.), este no es el caso para todos los comandos. En particular, las funciones para establecer límites, etiquetas y títulos se modifican ligeramente. Para realizar la transición entre las funciones de estilo MATLAB y los métodos orientados a objetos, realizamos los siguientes cambios (Puede consultar todos los atributos y métodos de la clase `axes` en este [link](https://matplotlib.org/3.1.1/api/axes_api.html#array)):

* `plt.plot()`   → `ax.plot()`
* `plt.legend()` → `ax.legend()`
* `plt.axis()`   → `ax.axis()`

* `plt.xlabel()` → `ax.set_xlabel()`
* `plt.ylabel()` → `ax.set_ylabel()`
* `plt.xlim()`   → `ax.set_xlim()`
* `plt.ylim()`   → `ax.set_ylim()`
* `plt.title()`  → `ax.set_title()`



Realicemos la misma gráfica anterior en el estilo orientado a objetos:

```
fig, ax = plt.subplots()

ax.plot(x, y, label="$sin(x)$")

# etiquetas de los ejes
ax.set_xlabel("$x$")
ax.set_ylabel("$sin(x)$");

# titulo
ax.set_title("Funcion seno");

# limites de los ejes
ax.axis("equal");

# leyenda
ax.legend();
```

En el estilo orientado a objetos, en lugar de llamar estas funciones individualmente, es más conveniente usar el método `ax.set()` para establecer todas estas propiedades a la vez:

```
fig, ax = plt.subplots()

ax.plot(x, y, label="$sin(x)$")

# etiquetas de los ejes y titulo
ax.set(xlabel="$x$",
       ylabel="$sin(x)$",
       title="Funcion seno")

# limites de los ejes
ax.axis("tight");

# leyenda
ax.legend();
```

# **Creando mútiples gráficos**

Para la generación de la figura y de uno o más conjuntos de ejes podemos recurrir a la función `plt.subplots`, la cual crea estos objetos y devuelve una referencia a la figura y al conjunto de ejes (si solo se ha creado uno) o a una lista de referencias a los conjuntos de ejes (si se han creado dos o más). 

Como ya vimos, la función `plt.subplots` crea por defecto una figura y un solo eje. Para tener dos gráficos en el mismo eje utilizamos la función `plot` reiteradamente: 

```
fig, ax = plt.subplots()

ax.plot(x, np.sin(x), label="$sin(x)$")
ax.plot(x, np.cos(x), label="$cos(x)$")

ax.set(xlabel="$x$",
       title="Funciones seno y coseno")

ax.legend();
```

Si queremos especificar el número de ejes a usar, debemos hacerlo explícitamente en la función `plt.subplots`. Sus dos primeros argumentos representan el número de filas y columnas a utilizar, solo que en este caso la referencia a las figuras se hace mediante `ax`:

```
fig, ax = plt.subplots(2,2)

print(type(ax))
print(ax.shape)
```

Note que `ax` es un arreglo de NumPy, en este caso de forma `(2,2)`, donde cada una de las entradas contiene un conjunto de ejes. Para pintar una gráfica en alguno de estos conjuntos de ejer simplemente debemos referenciar el eje mediante el elemento de `ax` correspondiente:

```
fig, ax = plt.subplots(2, 2, figsize=(14,7))

ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, 4*np.cos(x))
ax[1,1].plot(x, 4*np.sin(x))

ax[0,0].set(title="$sen(x)$", xlabel="x", ylabel="y")
ax[0,1].set(title="$cos(x)$", xlabel="x", ylabel="y")
ax[1,0].set(title="$4cos(x)$", xlabel="x", ylabel="y")
ax[1,1].set(title="$4sen(x)$", xlabel="x", ylabel="y");
```

```
# alternativamente
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2, 2, figsize=(14,7))

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, 4*np.cos(x))
ax4.plot(x, 4*np.sin(x))

ax1.set(title="$sen(x)$", xlabel="x", ylabel="y")
ax2.set(title="$cos(x)$", xlabel="x", ylabel="y")
ax3.set(title="$4sen(x)$", xlabel="x", ylabel="y")
ax4.set(title="$4cos(x)$", xlabel="x", ylabel="y");
```

Note que en este caso los nombres de las gráficas se están solapando con las etiquetas del eje $x$. Para tener cada atributo de la figura en su lugar, podemos utilizar el argumento por palabra clave `constrained_layout` en la función `plt.subplots`:

```
fig, ax = plt.subplots(2, 2, figsize=(14,7), constrained_layout=True)

ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, 4*np.cos(x))
ax[1,1].plot(x, 4*np.sin(x))

ax[0,0].set(title="$sen(x)$", xlabel="x", ylabel="y")
ax[0,1].set(title="$cos(x)$", xlabel="x", ylabel="y")
ax[1,0].set(title="$4cos(x)$", xlabel="x", ylabel="y")
ax[1,1].set(title="$4sen(x)$", xlabel="x", ylabel="y");
```

Otros argumentos por palabra clave útiles de la función `plt.subplots`, son `sharex` y `sharey`, que nos permiten compartir los ejes $x$ y $y$, respectivamente, entre las diferentes gráficas de diferentes formas.

También podemos añadir un título a la figura con el método `suptitle` del objeto `fig`.

```
# compartiendo todos los ejes
fig, ax = plt.subplots(2, 2, figsize=(14,7), constrained_layout=True, sharex=True, sharey=True)

fig.suptitle("Algunas funciones trigonometricas", fontsize=20)

ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, 4*np.cos(x))
ax[1,1].plot(x, 4*np.sin(x))

ax[0,0].set(title="$sen(x)$", ylabel="y")
ax[0,1].set(title="$cos(x)$", xlabel="x")
ax[1,0].set(title="$4cos(x)$", xlabel="x", ylabel="y")
ax[1,1].set(title="$4sen(x)$", xlabel="x");
```

```
# compartiendo los ejes x y compartiendo el eje y por fila
fig, ax = plt.subplots(2, 2, figsize=(14,7), constrained_layout=True, sharex=True, sharey="row")

fig.suptitle("Algunas funciones trigonometricas", fontsize=20)

ax[0,0].plot(x, np.sin(x))
ax[0,1].plot(x, np.cos(x))
ax[1,0].plot(x, 4*np.cos(x))
ax[1,1].plot(x, 4*np.sin(x))

ax[0,0].set(title="$sen(x)$", ylabel="y")
ax[0,1].set(title="$cos(x)$", xlabel="x")
ax[1,0].set(title="$4cos(x)$", xlabel="x", ylabel="y")
ax[1,1].set(title="$4sen(x)$", xlabel="x");
```

Recuerde consultar la documentación de la función `plt.subplots` para obtener más información respecto a los argumentos disponibles de la función.

# **Gráficos de dispersión**

Los gráficos de dispersión se utilizan para comparar los valores que toman dos variables distintas, una de las cuales se representa a lo largo del eje $x$ y la otra a lo largo del eje $y$. La gráfica resultante nos permite identificar visualmente la posible correlación entre las dos variables. 

Para realizar este tipo de gráficos podríamos utilizar la función `plt.plot` y remover las líneas. Sin embargo, Matplotlib ofrece una función mucho más potente `plt.scatter`. Esta función acepta como primeros dos argumentos los valores $x$ y $y$ de los puntos a mostrar. 

Hagamos un ejemplo sencillo: generemos dos variables que tengan un cierto grado de correlación:

```
# generamos un arreglo particular
x = np.linspace(0,50)

# definimos otra variable agregando
# "ruido" a la variable anterior 
y = x + 10*np.random.randn(50)

plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y");
```

El submódulo `random` utiliza un algoritmo particular, llamado Mersenne Twister, para generar números pseudoaleatorios.

Podemos definir una semilla (seed) que haga predecibles los números pseudoaleatorios. Si no asignamos la semilla, NumPy selecciona automáticamente un valor de semilla aleatorio basado en el dispositivo generador de números aleatorios del sistema o en el reloj. Cuando se establece un valor particular para la semilla, los mismos números aparecerán cada vez




```
# definiendo la semilla
np.random.seed(42)

x = np.linspace(0,50)
y = x + 10*np.random.randn(50)

plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y");
```

Definir una semilla es muy útil ya que nos permite tener procesos "aleatorios" que podamos reproducir.



Existen casos en los que los datos se presentan en un formato en los que podemos acceder a variables particulares mediante una palabra clave, por ejemplo, cuando tenemos los datos almacenados en un diccionario. En estos casos, debemos especificar la estructura de datos con el argumento por palabra clave `data`, y hacer refencia a las variables con su palabra clave como se muestra a continuación:

```
np.random.seed(42)

datos = {"x":np.linspace(0,50)}
datos["y"] = datos['x'] + 10*np.random.randn(50)

plt.scatter("x","y", data=datos)
plt.xlabel("x")
plt.ylabel("y");
```

Este tipo de gráfica resulta más útil cuando estamos visualizando elementos de diferente clase y los mostramos en el gráfico con diferenes colores. El color de cada punto se controla mediante el argumento por palabra clave `c`. Además, podemos cambiar el mapa de colores utilizando el argumento por palabra clave `cmap`. *Puede consultar los mapas de colores disponibles en esta [dirección](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)*

```
# generemos un arreglo para separar los datos en dos clases
clase = (datos["x"] < 25) + (datos["x"] > 25)*2
clase
```

utilizamos este arreglo para definir el color de cada una de las clases

```
plt.scatter("x","y", c=clase, data=datos, cmap="coolwarm")
plt.xlabel("x")
plt.ylabel("y");
```

Podemos además modificar el tamaño de cada uno de los puntos mediante el argumento por palabra clave ``s``:

```
# el tamaño de los puntos aumenta a medida que x aumenta
tamaño = np.arange(10,60)

plt.scatter("x","y", c=clase, s=tamaño, data=datos, cmap="coolwarm")
plt.xlabel("x")
plt.ylabel("y");
```

Podemos controlar qué tan opaco es el color con el argumento por palabra clave `alpha` que va de 0 (transparente) a 1 (opaco)

```
plt.scatter("x","y", data=datos, c=clase, s=tamaño, alpha=0.7, cmap="coolwarm")
plt.xlabel("x")
plt.ylabel("y");
```
